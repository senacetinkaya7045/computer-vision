# Computer Vision

Computer vision nedir? 
Computer Vision (Bilgisayar Görüşü), insan görsel sisteminin yapabileceklerini anlama ve otomatikleştirmede üst düzey bir anlayış kazanabilme ile ilgilenen disiplinler arası bir bilimsel alandır. Bilgisayar Görüşünün amacı, bilgisayarların olabildiğince insanlar gibi görmelerini yani yaşam içerisindeki sahnelere veya nesnelere anlam yüklemesini sağlamaktır. Bu bağlamda anlam yüklemek, görsel verilerden sembolik bilgilerin çözülmesi olarak görülebilir. Bilgisayarlar için bu, fotoğraflar ve videolar gibi görsel verilerdir. Bilgisayar Görüşü ilk olarak 1950’li yıllarda Ay’dan alınan görsel verileri anlamlandırmak için, 1990’lardan sonra ise görsel verileri daha estetik kılmak için kullanılmıştır. 2000’li yılların ortasında ise videoların yaygınlaşması ile hareketli nesneleri algılamak ve anlamlandırmak için kullanılmıştır.

Computer Vision (Bilgisayar Görüşü) Görevleri

Tanıma, ilk olarak tanıma görevi görüntü verilerinin bir nesne veya özellik içerme durumunu belirlemektedir. Tanıma görevinin alt başlıklarında ise nesne tanıma, tanımlama ve algılama görevleri bulunmaktadır.

İçeriğe Dayalı Görüntü Alma, oluşumunda farklı türden içerikler bulunduran geniş bir görüntü kümesinden ayrımlar yapıp istenilen görüntüleri elde etmeyi sağlar.

Poz Tahmini, belirlenen bir nesnenin çekim yapan kamera baz alınarak konumunu veya yönünü tahmin etmede kullanılır.

Hareket Analizi, bu görev üç ayrı başlıktan oluşmaktadır. 

Bunlar; Egomotion, İzleme ve Optik Akıştır.
Egomotion, döndürme ve çevirmeler sonucunda kamera ile üretilen bir görüntü kümesinden belirlemeler yapılmasıdır.
İzleme, görüntü kümesindeki hedef noktaların veya nesnelerin hareketlerinin takip edilmesidir.
Optik Akış ise görüntüde var olan her noktanın görüntü düzlemindeki hareketlerini belirlemek için kullanılmaktadır.

### 1- 

```python

#KAMERA İŞLEMLERİ
# Pencere adını belirler.
windowName = "KAMERA"
#Belirtilen adla bir pencere oluşturur.
cv2.namedWindow(windowName)
#Kamerayı yakalamak için bir VideoCapture nesnesi oluşturur.
# 0, var olan  kamerayı temsil eder.
cap = cv2.VideoCapture(0)
#Çemberin rengini belirler.
color = (0, 255, 255)
#Çemberin çizgi kalınlığını belirler.
line_width = 3
#Çemberin yarıçapını belirler.
radius = 200
#Başlangıçta çemberin merkez noktasını belirler. 
# Bu, fare etkileşimiyle güncellenecek bir değişken olarak atanır.
point = (0, 0)


#Fare etkileşimlerini izlemek için bir fonksiyon tanımlanır.
# Bu fonksiyon, fare hareket edildiğinde merkez noktayı güncellemek için kullanılacaktır.
def touch(event, x, y, flags, param):
    global point
    if event == cv2.EVENT_MOUSEMOVE:
        point = x, y

# Fare etkileşimlerini "KAMERA" penceresine bağlar ve 
# bu etkileşimlerin işlenmesi için tanımlanan touch fonksiyonunu çağırır.
cv2.setMouseCallback(windowName, touch)

#Sonsuz bir döngü başlatır.
while True:
    #Kameradan bir kare yakalar ve bu kare video değişkenine atanır.
    ret, video = cap.read()
    #Yakalanan kare üzerine bir çember çizer. 
    # Çemberin merkezi, yarıçapı, rengi ve kalınlığı belirtilir.
    cv2.circle(video, point, radius, color, line_width)
    #Çizilen çemberi göstermek için "KAMERA" penceresini günceller.
    cv2.imshow(windowName, video)
    #Kullanıcının 's' tuşuna basıp basmadığını kontrol eder.
    if cv2.waitKey(1) == ord('s'):
        break

#Kullanıcının 's' tuşuna basıp basmadığını kontrol eder.
cap.release()
#Tüm OpenCV pencerelerini kapatır.
cv2.destroyAllWindows()

```

### 3- 

```python
# RESİM DÖNDERME
import cv2
import numpy as np
image= cv2.imread("manzara.jpg",1)
#Görüntünün döndürülmesi için bir dönüşüm matrisi oluşturur.
# (150,150) dönüşümün merkezini belirtir. -30 dönüş açısını belirtir (saat yönünün tersine, negatif yönde döner). 
# Son argüman olan 1, dönüşümün ölçeklendirme yapmadığını ifade eder.
rot = cv2.getRotationMatrix2D((150,150), -30,1)
#cv2.warpAffine fonksiyonunu kullanarak görüntüyü döndürür.
# İlk argüman olarak orijinal görüntüyü, ikinci argüman olarak dönüşüm matrisini ve 
# üçüncü argüman olarak döndürülmüş görüntünün boyutunu alır.
image_rot = cv2.warpAffine(image,rot,(image.shape[0], image.shape[1]))
cv2.imshow("Rot",image_rot)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

# RESİM BULANIKLAŞTIRMA
import cv2
import numpy as np
image= cv2.imread("manzara.jpg",1)
#Gaussian bulanıklaştırma işlemi uygular. İlk argüman olan image, 
# işlenecek olan resmi temsil eder. İkinci argüman, bulanıklaştırma için kernel boyutunu belirtir. 
# Burada (1, 1) kullanıldığından, çok hafif bir bulanıklık oluşur
blur = cv2.GaussianBlur(image,(1,1),1)
#bu sefer kernel boyutu (55, 55) olarak belirlenir, bu da daha yoğun bir bulanıklık oluşturur.
blur_2 = cv2.GaussianBlur(image, (55,55), 1)
#bulanıklaştırılmış resmi göstermek için bir pencere oluşturur. 
cv2.imshow("Blur",blur)
#Bu satırda da diğer bulanıklaştırılmış resmi göstermek için bir pencere oluşturulur.
cv2.imshow("Blur2",blur_2)
#bir tuşa basılana kadar beklemeyi sağlar. 
cv2.waitKey(0)
#Bu satır, tüm OpenCV pencerelerini kapatır ve programı sonlandırır.
cv2.destroyAllWindows()
```

```python
# NESNE AYIKLAMA 
#simple thresholding işlevini uygulayacağız. Bunu uygulamak için okuma yaptığımız resmi siyah-beyaz modda okuyacağız. 
# Daha sonra bir eşik değeri belirterek belli kısımları daha koyu yaparak yada belli kısımları beyaz yaparak yok edeceğiz. 
# Bu sayede nesnelerimizi ortaya çıkaracağız
import cv2
import numpy as np
img = cv2.imread("kiraz.jpeg",0)
img = cv2.resize(img,(500,500))
#cv2.adaptiveThreshold kodu ile nesneleri ayırt etmek için bir eşikleme işlemi uygulanır
thresh = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,1001,0)
#cv2.threshold(src(RESMİ ATADIĞIMIZ DEĞİŞKEN), thresh(EŞİK DEĞER), maxval(PİKSEL), type(AYIKLAMA TİPİMİZ))
ret,thresh_2=cv2.threshold(img, 190, 255, cv2.THRESH_BINARY)
thresh_3 = np.concatenate((img,thresh,thresh_2),axis=1)
cv2.imshow("KAMERA",thresh_3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

# NESNE AYIKLAMA VE SINIRLARINI ÇİZME 
import cv2
import numpy as np
#resmi renkli modda okue
img = cv2.imread("elmalar.png",1)
#resmi gri moda çevirir
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
thresh=cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 10001, 1)
#çizeceğimiz sınırları cv2.findContours() fonksiyonu ile buluyoruz
contours,hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
index = 0
#hickness ise çizdiğimiz sınırların kalınlığı 
thickness = 1
#color contour rengi bgr modda ayarlanır
color=(255,0,255)
cv2.drawContours(img, contours, index, color,thickness)
cv2.imshow("KAMERA",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

# İNSAN YÜZ HATLARINI BELİRLEME
import cv2
import numpy as np
img = cv2.imread("face.jpg",1)
img=cv2.resize(img,(500 ,470))
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
# Adaptif eşikleme yöntemini kullanarak gri tonlamalı resimde eşikleme işlemi uygulanır
thresh=cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 401, 1)
#Eşiklenmiş resimdeki kontur noktalarını bulur. contours değişkeni, kontur noktalarının koordinatlarını içerir.
contours,hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
#Çizilen konturların indeksini belirler. -1, tüm konturların çizilmesini sağlar.
index = -1
#Kontur çizgilerinin kalınlığını belirler.
thickness = 1
#Kontur çizgilerinin rengini belirler.
color=(255,20,20)
#Orijinal resmi kopyalar.
a=img.copy()
#Resmin üzerine konturları çizer.
cv2.drawContours(img, contours, index, color,thickness)
# İki resmi yan yana birleştirir. 
final=np.concatenate((a,img),axis=1)
#İki resmin birleştirilmiş halini "output.png" adlı bir dosyaya kaydeder.
cv2.imwrite("output.png",final)
#Birleştirilmiş resmi gösterir.
cv2.imshow("KAMERA",final)
# Bir tuşa basılana kadar bekler.
cv2.waitKey(0)
#Tüm OpenCV pencerelerini kapatır.
cv2.destroyAllWindows()
```

```python
# KENAR ALGILAMA VE FİLTRELEME 
import cv2
import numpy as np

# Görüntüyü oku
image = cv2.imread("indir.jpeg")


# Görüntü boyutlarını al
height, width = image.shape[:2]

# Sobel kenar algılama uygula
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=5)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=5)

# Sobel sonuçlarını görselleştir
cv2.imshow("Sobel X", sobel_x)
cv2.waitKey(0)
cv2.imshow("Sobel Y", sobel_y)
cv2.waitKey(0)

# Sobel sonuçlarını birleştir
sobel_OR = cv2.bitwise_or(sobel_x, sobel_y)
cv2.imshow("Sobel OR", sobel_OR)
cv2.waitKey(0)

# Laplacian kenar algılama uygula
laplacian = cv2.Laplacian(image, cv2.CV_64F)
cv2.imshow("Laplacian", laplacian)
cv2.waitKey(0)

# Canny kenar algılama uygula
canny = cv2.Canny(image, 50, 120)
cv2.imshow("Canny", canny)
cv2.waitKey(0)

# Tüm pencereleri kapat
cv2.destroyAllWindows()
```

```python
# RESİM OKUMA VE KAYDETME 
import cv2
import numpy as np
#cv2.imread( ) fonksiyonununun içerisinde ilk kısma çift tırnak içerisinde okunacak resmimizin
# ismini uzantısı ile beraber yazarken, ikinci kısma ise -1,0,-1 sayılarından birini yazıyoruz.
#1 = cv2.IMREAD_COLOR = Resimleri renkli olarak okur.
#0 = cv2.IMREAD_GRAYSCALE = Resimleri gri modda okur.
#1 = cv2.IMREAD_UNCHANGED = Resimleri alfa modda okur. 
img = cv2.imread("ata.png",1)
#resmimizi göstermek için bir pencere oluşturuyoruz. Bu pencereyi oluşturmak için 
# cv2.namedWindow() fonksiyonunu kullanıyoruz.
windowName = "Output Screen"
cv2.namedWindow(windowName,0)
#bu oluşturduğumuz pencere göstermek istediğimiz resmi
# cv2.imshow() fonksiyonu ile çağırıyoruz
cv2.imshow(windowName, img)
#bu resmimizi dışarıya bir dosya olarak çıktı almak adına
# cv2.imwrite() fonksiyonunu çağırıyoruz.
cv2.imwrite("Output.jpg",img)
#cv2waitkey()fonksiyonu içerisine yazacağınız sayı ile pencerenin ne kadar açık kalacağı belirlenir
# eğer 0 yazarsak herhengi bir tuşa basılana kadar açık kalacaktır
cv2.waitKey(0)
#cv2.destroyWindow() fonksiyonu içerisine
# hangi pencereyi kapatacağımızı yazarak kodumuzu tamamlıyoruz.
cv2.destroyWindow(windowName)
```
